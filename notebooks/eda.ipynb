```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Enable inline plotting
%matplotlib inline

# Load the dataset
df = pd.read_csv("../data/student-mat.csv", sep=";")
```

```python
print("Dataset shape:", df.shape)
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Basic data types
df.info()
```

```python
# Summary of numeric columns
df.describe()
```

```python
# Distribution of final grades
sns.histplot(df["G3"], kde=True)
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Frequency")
plt.show()
```

```python
# Gender distribution
sns.countplot(x="sex", data=df)
plt.title("Student Gender Distribution")
plt.show()

# Internet access and study time
sns.countplot(x="internet", hue="studytime", data=df)
plt.title("Study Time by Internet Access")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include="number")
correlation = numeric_cols.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# G1 and G2 vs G3 (first and second period grades vs final)
sns.scatterplot(x="G1", y="G3", data=df, alpha=0.6)
plt.title("First Period Grade vs Final Grade")
plt.show()

sns.scatterplot(x="G2", y="G3", data=df, alpha=0.6)
plt.title("Second Period Grade vs Final Grade")
plt.show()
```

```python
# Study time vs G3
sns.boxplot(x="studytime", y="G3", data=df)
plt.title("Study Time vs Final Grade")
plt.show()
```

## ðŸ” Key Insights

- G1 and G2 (previous grades) are highly correlated with G3 â€” strong predictors.
- Study time shows a general upward trend in performance.
- Internet access doesn't have a strong visual correlation with performance.
- No missing data â€” good for model training.
- Some features (e.g., absences, failures) may need normalization or transformation.
